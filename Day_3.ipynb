```python
#lists and dictionaries \

# mutable and unmutable
# mutable (THAT CAN BE CHANGED) #unmutable (THAT CANNOT BE CHANGED OR MODIFIED)

#index counting works here

list_of_numbers = [1,2,"tanvi",4.5,6.2]
#print(list_of_numbers)

list_of_numbers.append("rohit")
#print(list_of_numbers)

list_of_numbers[3] = "sharukh"
#print (list_of_numbers)

list_of_numbers.insert(1,"hello")
#print (list_of_numbers)

print(list_of_numbers.index(6.2))
```

```python

#creating a TUPLE - TUPLE IS IMMUTABLE
numbers = (9,8,7)
print(numbers)

numbers[4] = 6
print(numbers)

#error because tuples is immutable
```

```python
# key- value pairs
# you can modify after creating (mutable)

car_details = {
    "brand" : "tata",
    "model" : "harrier",
    "color" : "white",
    "year_of_manufacture" : 2022
}
car_details["milage"] = 18

print(car_details)

#print(car_details)

#print(car_details['year_of_manufacture'])

for item, key in car_details.items():
  print(f"{item} : {key}")
```

```python
file_name = "example.txt"
data_to_push = ["example sentence to print in a file", "this is the second line"]

try:
  with open(file_name, 'w') as file:
    file.writelines(line + "\n" for line in data_to_push)
  print("data is written in the file")
except IOError as e:
  print(e)
```
