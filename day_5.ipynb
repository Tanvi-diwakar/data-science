MEAN, MEDIAN AND MODE

```python
mean_weather_data = np.mean(weather_data, axis=1)   #Average across months -- row-wise
print("the mean temperature of the dataset is ", mean_weather_data)
```

```python
# trying to find hottest and coldest city
hottest_month = np.argmax(weather_data, axis=1)
print("the hottest month per city: ", hottest_month)
coldest_month = np.argmin(weather_data, axis=1)
print("the coldest month per city: ", coldest_month)
```

```python
# to get direct temperature value
hottest_month_values = np.max(weather_data, axis=1)
print("the hottest month per city: ", hottest_month_values)
coldest_month_values = np.min(weather_data, axis=1)
print("the coldest month per city: ", coldest_month_values)
```
